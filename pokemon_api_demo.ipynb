```python
# Step 1: Import the requests library
import requests
import pandas as pd
```

```python
# Step 2: Choose a PokÃ©mon to explore
pokemon_name = "pikachu"
```

```python
# Step 3: Make a request to the PokeAPI
url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}"
response = requests.get(url)
response
```

```python
# Step 4: Parse the JSON response
data = response.json()
type(data)
```

```python
# Step 5: Extract characteristics
name = data['name']
height = data['height']
weight = data['weight']
types = [t['type']['name'] for t in data['types']]
abilities = [a['ability']['name'] for a in data['abilities']]
```

```python
# Step 6: Create a DataFrame for easier display and further use
df = pd.DataFrame([{
    'name': name,
    'height': height,
    'weight': weight,
    'types': ", ".join(types),
    'abilities': ", ".join(abilities)
}])

df
```

```python
# Step 7: Export to csv
df.to_csv("outputs/pokemon_info.csv", index=False)
```
